```python
import pandas as pd
from pathlib import Path

batt_park = pd.read_csv('Batt_park_2022_sales.csv')
other_three_df = pd.read_csv('highest_three_lowest_four.csv')

frames = [batt_park, other_three_df]
highest_and_lowest_df = pd.concat(frames)
```

```python
highest_and_lowest_df.drop(columns = ['Purchase Date', 'Purchase Type', 'City', 'Listed for Sale?', 'CLTV %'], inplace = True)
highest_and_lowest_df.head()
```

```python
highest_and_lowest_df[['1st Amount', '2nd Amount']] = highest_and_lowest_df[['1st Amount', '2nd Amount']].fillna(0)
highest_and_lowest_df.head()
```

```python
highest_and_lowest_df["Mortgage Amount"] = highest_and_lowest_df['1st Amount']+ highest_and_lowest_df['2nd Amount']
highest_and_lowest_df.drop(columns = ['1st Amount', '2nd Amount'], inplace = True)
investor_hl_df = highest_and_lowest_df[(highest_and_lowest_df['Owner Occ?'] ==1)]
highest_and_lowest = highest_and_lowest_df[(highest_and_lowest_df['Owner Occ?'] == 0)]
```

```python
sales_count = highest_and_lowest['ZIP'].count()
total_sales = highest_and_lowest['Purchase Amt'].sum()
ave_dp = highest_and_lowest['Purchase Down %'].mean().round(1)
total_mortgaged = highest_and_lowest['Mortgage Amount'].sum()
ave_sale = highest_and_lowest['Purchase Amt'].mean().astype('int64')
ave_mtg = highest_and_lowest['Mortgage Amount'].mean().round(0)

summary_df = pd.DataFrame({'Total Sales 2022': [sales_count], 'Combined Sales': [total_sales],
                           'Average DownPayment': [ave_dp], 
                           'Total Mortgaged': [total_mortgaged], 
                           'Average Mortgage': [ave_mtg], 
                           'Average Sales Price': [ave_sale],})


# ({'ZIP': 'count', 'Purchase Amt': 'sum', 'Purchase Down %': 'mean', 'Mortgage Amount': 'sum'}, index = 0)
# batt_park_summary

summary_df
```

```python
summary_df['Mtg Tx pd'] = summary_df['Total Mortgaged']*0.018
summary_df['Est Sale Tax'] = summary_df['Combined Sales']*0.009
summary_df
```

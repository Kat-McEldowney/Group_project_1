```python
batt_park_df.drop(columns = ['Purchase Date', 'Purchase Type', 'City', 'Listed for Sale?', 'CLTV %'], inplace = True)
batt_park_df.head()
```

```python
batt_park_df[['1st Amount', '2nd Amount']] = batt_park_df[['1st Amount', '2nd Amount']].fillna(0)
batt_park_df.head()
```

```python
batt_park_df["Mortgage Amount"] = batt_park_df['1st Amount']+ batt_park_df['2nd Amount']
batt_park_df.drop(columns = ['1st Amount', '2nd Amount'], inplace = True)
batt_park_df.head()
```

```python
sales_count = batt_park_df['ZIP'].count()
total_sales = batt_park_df['Purchase Amt'].sum()
ave_dp = batt_park_df['Purchase Down %'].mean().round(1)
total_mortgaged = batt_park_df['Mortgage Amount'].sum()
ave_sale = batt_park_df['Purchase Amt'].mean().astype('int64')
ave_mtg = batt_park_df['Mortgage Amount'].mean().round(0)

summary_df = pd.DataFrame({'Total Sales 2022': [sales_count], 'Combined Sales': [total_sales],
                           'Average DownPayment': [ave_dp], 
                           'Total Mortgaged': [total_mortgaged], 
                           'Average Mortgage': [ave_mtg], 
                           'Average Sales Price': [ave_sale],})


# ({'ZIP': 'count', 'Purchase Amt': 'sum', 'Purchase Down %': 'mean', 'Mortgage Amount': 'sum'}, index = 0)
# batt_park_summary

summary_df
```

```python
summary_df['Mtg Tx pd'] = summary_df['Total Mortgaged']*0.018
summary_df['Est Sale Tax'] = summary_df['Combined Sales']*0.009
summary_df
```
